```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
tensor1 =  torch.linspace(-1.5,1.5,2000)
support = 1
a = tensor1>support
b = tensor1<-support
c = torch.logical_not(torch.logical_or(a, b))
out = a * 1.0 + b * 1.0 + c * (-torch.cos(2 * torch.pi/support/2 * tensor1)+1)/2 
plt.plot(tensor1,out)
plt.grid(True)
```
