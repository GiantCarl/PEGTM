```python
import numpy as numpy
import torch
import matplotlib.pyplot as plt
```

```python
tensor1 =  torch.linspace(-0.2,0.2,2000)
phase_field_variable = 0.01
```

```python
out = torch.exp(-torch.abs(tensor1)/phase_field_variable)
```

```python
plt.plot(tensor1,out)
```

```python
a = tensor1>0
b = tensor1<0
derivate = -1*a/phase_field_variable *torch.exp(-tensor1/phase_field_variable) + b/phase_field_variable*torch.exp(tensor1/phase_field_variable)
```

```python
plt.plot(tensor1,derivate)
```
